```python
##availableDts
##availableDts is string contains 4 integer in the form 1,2,3,4
#1 is the available dates in July
#2 is the available dates in August
#3 is the available dates in September
#4 is the available dates in November


##since I scrapped the data in August 18
##we will seperate #2, #3, #4 in 3 columns to show the available dates for each months

listingDt['availableDtsAug'] = (listingDt['availableDts'].str.split(',', expand = True)[1]).astype(float)
listingDt['availableDtsSep'] = (listingDt['availableDts'].str.split(',', expand = True)[2]).astype(float)
listingDt['availableDtsNov'] = (listingDt['availableDts'].str.split(',', expand = True)[3]).astype(float)

(listingDt[['availableDts','availableDtsAug', 'availableDtsSep','availableDtsNov']]).head()
```

```python
##amendities
##amendities is string contains all the amendities available

for x in ((listingDt['amendities'])[1:2]).str.split('\n'):
    print(x )
    print('\n')
    
##This string includes both amendities and comments
##to remove comments, remove string that have >= 5 words 

##Amentities Listings has about 9 different subgroups
##Basic
##Family features
##Facilities
##Dining
##Guest access
##Logistics
##Bed and bath
##Safety features
##Not included

##seperate these 9 subgroups into different columns

###Step1: filter out string with >= 5 words, those are highly likely to be comments
def removeComments(amenityString):
#     print(amenityString)
    amenities = []
    for x in amenityString.split('\n'):
        wordCounts = len(x.split(' '))
        if wordCounts < 5:
            amenities.append(x)
     
    amenitiesString = '\n'.join(str(elem) for elem in amenities)
    return amenitiesString

###Step2: Break amenity string into subgroups
def getSubAmenities(amenityString):
    subAmenities = ['Basic', 'Family features', 'Facilities', 'Dining', 'Guest access', 'Logistics', 'Bed and bath', 'Safety features', 'Not included']
    ##get substring that exists in amenityString
    subAmenitiesExist = [x for x in subAmenities if x in amenityString]
#     print(*subAmenitiesExist)
    
    subAmenitiesStrings = []
    for x in subAmenities:
        if not x in subAmenitiesExist:
            subAmenitiesStrings.append(None)
        else:
            startIndex = int(amenityString.index(x))
#             print(startIndex)
            ##find the index of the element after x in subAmenitiesExist
            indexXNext = subAmenitiesExist.index(x) + 1
            if indexXNext == len(subAmenitiesExist): #x is the last element
                endIndex = int(len(amenityString))
            else:
                endIndex = int(amenityString.index(subAmenitiesExist[indexXNext]) - 1)
#             print(endIndex)
            AmenitiesStr = amenityString[startIndex: endIndex]
            subAmenitiesStrings.append(AmenitiesStr)
    subAmenitiesStrings = ','.join(str(elem) for elem in subAmenitiesStrings)
    return subAmenitiesStrings


##put step1, step2 together
def amenityProcessing(amenityString):
    amenityStep1 = removeComments(amenityString)
#     print(amenityStep1)
    amenityStep2 = getSubAmenities(amenityStep1)
    return amenityStep2


##Apply step1, step2 to the whole dataframe
listingDt['amenityString'] = listingDt['amendities'].astype(str).apply(amenityProcessing)

listingDt['amenityBasic'] = (listingDt['amenityString'].str.split(',', expand = True)[0]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityFamily'] = (listingDt['amenityString'].str.split(',', expand = True)[1]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityFacilities'] = (listingDt['amenityString'].str.split(',', expand = True)[2]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityDining'] = (listingDt['amenityString'].str.split(',', expand = True)[3]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityAccess'] = (listingDt['amenityString'].str.split(',', expand = True)[4]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityLogistics'] = (listingDt['amenityString'].str.split(',', expand = True)[5]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityBB'] = (listingDt['amenityString'].str.split(',', expand = True)[6]).str.split('\n', n=1, expand = True)[1]
listingDt['amenitySafety'] = (listingDt['amenityString'].str.split(',', expand = True)[7]).str.split('\n', n=1, expand = True)[1]
listingDt['amenityNotIn'] = (listingDt['amenityString'].str.split(',', expand = True)[8]).str.split('\n', expand = True)[2]

listingDt[['amenityString', 'amenityBasic', 'amenityFamily', 'amenityFacilities', 'amenityDining', 'amenityAccess',
          'amenityLogistics', 'amenityBB', 'amenitySafety', 'amenityNotIn']].head()
```

```python
###Mapping listingHouseType
HouseTypeMapping = dict.fromkeys(['Boat', 'Houseboat'] , 'Houseboat')
HouseTypeMapping.update(dict.fromkeys(['Camper/RV', 'Tent'] , 'Camper/RV/Tent'))
HouseTypeMapping.update(dict.fromkeys(['Entire apartment', 'Entire condominium', 'Entire guest suite', 'Entire loft', 'Entire serviced apartment'] , 'Entire Apartment'))
HouseTypeMapping.update(dict.fromkeys(['Entire bungalow', 'Entire cabin', 'Entire cottage', 'Entire guesthouse', 'Entire house', 'Entire in-law', 'Entire townhouse', 'Tiny house'] , 'Entire House'))

HouseTypeMapping.update(dict.fromkeys(['Private room in apartment', 'Private room in condominium', 'Private room in guest suite', 'Private room in serviced apartment'] , 'Private Room in Apartment'))
HouseTypeMapping.update(dict.fromkeys(['Private room in bungalow', 'Private room in cabin', 'Private room in cottage', 'Private room in dome house','Private room in guesthouse', 'Private room in house', 'Private room in townhouse', 'Private room in villa'] , 'Private Room in House'))
HouseTypeMapping.update(dict.fromkeys(['Private room in bed and breakfast', 'Room in hotel', 'Private room in casa particular (cuba)'] , 'Private Room in Hotel'))


listingDt["listingHouseTypeGroup"] = listingDt["listingHouseType"].map(HouseTypeMapping).fillna(listingDt["listingHouseType"])
                        
# listingHouseTypeCross = pd.crosstab(index=listingDt["listingHouseTypeGroup"], 
#                            columns=listingDt["listingHouseType"])

# listingHouseTypeCross  
```

```python
##Cleanup numBaths
###Mapping listingHouseType
numBathMapping = dict.fromkeys(['0 baths', '0 shared baths'] , '0 Baths')
numBathMapping.update(dict.fromkeys(['1 bath', '1 private bath'] , '1 Bath'))
numBathMapping.update(dict.fromkeys(['Half-bath'] , '0.5 Bath'))

listingDt["numBathGroup"] = listingDt["numBaths"].map(numBathMapping).fillna(listingDt["numBaths"])
listingDt["numBathNew"] = (listingDt['numBathGroup'].str.split(' ', n = 1, expand = True)[0]).astype(float)
listingDt["sharedBath"] = np.where(listingDt['numBathGroup'] == '0 Baths', -1, np.where(listingDt['numBathGroup'].str.contains('shared'), 1, 0))

bathNumCross = pd.crosstab(index=listingDt["numBathNew"], 
                           columns=listingDt["numBathGroup"])

bathSharedCross = pd.crosstab(index=listingDt["sharedBath"], 
                           columns=listingDt["numBathGroup"])

bathNumCross 
bathSharedCross
```
